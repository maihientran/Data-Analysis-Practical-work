```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson
mu = 400
nmax = 475
n = np.arange(nmax+1)

pdf_Poisson = poisson.pmf(n,mu)

xmin,xmax = 325,float(nmax)
ymin,ymax = 0.,0.025

fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(n, pdf_Poisson,marker='o',markersize=10,color='red',lw=3, )

ax.axvline(x = np.round(mu), color='red',linestyle='--', label=r'$\mu$ ='+str(mu) )   ## Vertical lines

ax.set_xlabel(r'$n$')
ax.set_ylabel(r'$P_{\rm Poisson}(n)$')
ax.set_xlim(xmin,xmax)
ax.set_ylim(ymin,ymax)

ax.legend()
ax.set_title(' Poisson distribution',fontsize=24,pad=15)
plt.savefig("Poisson_distribution.png",facecolor='w', dpi=600)
```
